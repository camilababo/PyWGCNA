# Compare agreement between filtered and unfiltered datasets

```python
import PyWGCNA
pyWGCNA_728_nofilt = PyWGCNA.readWGCNA("E-MEXP-728-non-filtered-TPM6.p")
pyWGCNA_728_filt = PyWGCNA.readWGCNA("E-MEXP-728-filtered-TPM1.p")
pyWGCNA_1304_nofilt = PyWGCNA.readWGCNA("E-MEXP-1304-non_filtered-TPM7.p")
pyWGCNA_1304_filt = PyWGCNA.readWGCNA("E-MEXP-1304-TPM1.p")
```

## Comparision of the E-MEXP-728 (filtered and non-filtered)

```python
module = pyWGCNA_728_filt.getGeneModule("black")
print(module)
```

```python
def save_locus_tags_to_file(module, output_file_path, module_name):
    # Extract locus tags from the module dictionary
    all_locus_tags = []

    for color_group, df in module.items():
        locus_tags = df.index.tolist()
        all_locus_tags.extend(locus_tags)

    # Save locus tags to the specified txt file
    with open(output_file_path, "a") as f:
        for locus_tag in all_locus_tags:
            f.write(f"{locus_tag} {module_name}\n")

    print(f"Locus tags for {module_name} saved to {output_file_path}")

# Example usage

module_list_728_filt = ["black", "lightcoral", "brown", "gainsboro", "rosybrown", "darkgrey", "white", "whitesmoke", "indianred", "silver", "darkred", "maroon", "red", "firebrick", "lightgrey"]

module_list_728_unfilt = ["lightcoral", "blanchedalmond", "mistyrose", "cornsilk", "gold", "linen", "sienna", "chocolate", "oldlace", "darkgoldenrod", "firebrick", "floralwhite", "brown", "black", "coral", "orange", "tomato", "darkred", "palegoldenrod"]

module_1304_filt = ["dimgrey", "darkgrey", "lightgrey"]

module_1304_unfilt = ["darkred", "indianred", "darksalmon", "white", "black", "mistyrose", "darkgrey", "lightcoral", "sienna", "rosybrown", "gainsboro", "lightsalmon"]

module_lists = [module_list_728_filt, module_list_728_unfilt, module_1304_filt, module_1304_unfilt]

list_pywgcna_objcts = [pyWGCNA_728_filt, pyWGCNA_728_nofilt, pyWGCNA_1304_filt, pyWGCNA_1304_nofilt]

for pyWGCNA_obj, module_lists in zip(list_pywgcna_objcts, module_lists):
    for module_name in module_lists:
        module = pyWGCNA_obj.getGeneModule(module_name)
        save_locus_tags_to_file(module, f"{pyWGCNA_obj.name}.txt", module_name)
```

```python
comparison = PyWGCNA.compareNetworks(PyWGCNAs = [pyWGCNA_728_nofilt, pyWGCNA_728_filt])
```

```python
comparison.jaccard_similarity
```

```python
comparison.fraction
```

```python
color = {"E-MEXP-728-non-filtered-TPM6": "lightblue",
         "E-MEXP-728-filtered-TPM1": "lightgreen"}
comparison.plotJaccardSimilarity(color=color,
                                 cutoff=0.01,
                                 plot_format="pdf",
                                 file_name="jaccard_similarity_728")
```

## Comparison between E-MEXP-1304 (filtered and un-filtered)

```python
comparison_1304 = PyWGCNA.compareNetworks(PyWGCNAs = [pyWGCNA_1304_nofilt,pyWGCNA_1304_filt])
comparison_1304.jaccard_similarity
```

```python
color_1304 = {"E-MEXP-1304-non_filtered-TPM7": "lightcoral",
             "E-MEXP-1304-TPM1": "lightpink"}
comparison_1304.plotJaccardSimilarity(color=color_1304,
                                 cutoff=0.05,
                                 plot_format="pdf",
                                 file_name="jaccard_similarity_1304")
```
