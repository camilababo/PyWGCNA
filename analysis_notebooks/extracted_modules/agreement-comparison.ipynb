# Comparison of extracted modules
Assessment of the Jaccard similarity index for all extracted modules with the PyWGCNA and BADDADAN approach.

```python
# Import developed method
from jaccard_similarity import CompareModulesBetweenApproaches
```

## Compare PyWGCNA

```python
# Import modules to compare
module_paths = ["E-MEXP-1304-non_filtered-TPM7.txt", "E-MEXP-728-non-filtered-TPM6.txt"]
comparison = CompareModulesBetweenApproaches(module_paths)
```

```python
# Create combined module file
comparison.create_combined_module_file(remove_non_overlapping_genes=False)
```

```python
# Calculate Jaccard similarity
comparison.calculate_jaccard_similarity()
```

```python
# Plot Jaccard similarity
comparison.plot_jaccard_similarity(cutoff=0.05, network_colors={"E-MEXP-1304-non_filtered-TPM7": "green", "E-MEXP-728-non-filtered-TPM6": "orange"})
```

## Plot progression of modules' correlation value with temperature (PyWGCNA)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data
data = pd.read_csv('../moduleTraitCor-E-MEXP-728-non-filtered-TPM6.csv', index_col=0)  # Replace with your data file

# Sort data based on correlation values
sorted_data = data.sort_values(by='Temperature')

# Extract module names and correlation values
module_names = sorted_data.index
correlation_values = sorted_data['Temperature']

# Plot the correlation progression and the trend line for each module
plt.figure(figsize=(10, 6))

plt.plot(module_names, correlation_values, marker='o', linestyle='-', label='Correlation')

plt.xlabel('Module')
plt.ylabel('Correlation')
plt.title('Correlation Progression for E-MEXP-728 Modules')
plt.xticks(rotation=45, ha='right')

plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
