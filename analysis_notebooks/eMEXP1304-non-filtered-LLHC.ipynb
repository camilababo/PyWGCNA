# Functional Enrichment Analysis

```python
geneList = PyWGCNA.getGeneList(dataset='athaliana_eg_gene',
                               attributes=['ensembl_gene_id',
                                           'external_gene_name',
                                           'gene_biotype'],
                               maps=['gene_id', 'gene_name', 'gene_biotype'])
```

```python
PyWGCNA_1304_nofilt.updateGeneInfo(geneList)
```

```python
PyWGCNA_1304_nofilt.figureType = "png"
gene_set_library = ["GO_Biological_Process_2021", "GO_Cellular_Component_2021", "GO_Molecular_Function_2021"]
PyWGCNA_1304_nofilt.functional_enrichment_analysis(type="GO",
                                             moduleName="gainsboro",
                                             sets=gene_set_library,
                                             p_value=0.00000005,
                                             file_name="GO_Dimgrey_2021")
```

```python
PyWGCNA_1304_nofilt.figureType = "png"
gene_set_library = ["KEGG_2019_Mouse"]
PyWGCNA_1304_nofilt.functional_enrichment_analysis(type="KEGG",
                                             moduleName="gainsboro",
                                             sets=gene_set_library,
                                             p_value=0.00000005)
```

## Module Comparison

```python
PyWGCNA_1304_nofilt_TPM0 = PyWGCNA.readWGCNA("E-MEXP-1304-TPM0.p")
PyWGCNA_1304_nofilt_TPM1 = PyWGCNA.readWGCNA("E-MEXP-1304-TPM1.p")
PyWGCNA_1304_nofilt_TPM2 = PyWGCNA.readWGCNA("E-MEXP-1304-TPM2.p")
PyWGCNA_1304_nofilt_TPM5 = PyWGCNA.readWGCNA("E-MEXP-1304-TPM5.p")
PyWGCNA_1304_nofilt_TPM6 = PyWGCNA.readWGCNA("E-MEXP-1304-TPM6.p")
```

```python
comparison = PyWGCNA.compareNetworks(PyWGCNAs=[PyWGCNA_1304_nofilt_TPM0,
                                                PyWGCNA_1304_nofilt_TPM1,
                                                PyWGCNA_1304_nofilt_TPM2,
                                                PyWGCNA_1304_nofilt_TPM5,
                                                PyWGCNA_1304_nofilt_TPM6])
```

## Functional Enrichment Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Read the dataset
df = pd.read_csv("fea-1304.csv", header=0, sep="\t")
df
```

```python
# Scale fold enrichment values for bubble size
scale_factor = 100
df['bubble_size'] = df['fold_enrichment'] * scale_factor

# Create the bubble plot
plt.figure(figsize=(50, 6))

# Create the bubble plot
plt.figure(figsize=(10, 8))
sns.set_style("whitegrid")

# Customize the bubble plot
bubble_plot = sns.scatterplot(
    y='biological_process',
    x='module_name',
    hue='p_value',  # Color based on biological process
    size='bubble_size',  # Size based on fold enrichment
    sizes=(50, 500),  # Adjust bubble size range as needed
    data=df,
)

# Add labels to axes
bubble_plot.set(ylabel='Biological Process', xlabel='Gene Module')

# Customize legend
bubble_plot.legend(title='Caption', loc='upper right')

# Show the plot
plt.title('Functional Enrichment Analysis')
plt.show()
```
