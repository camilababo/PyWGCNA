## Setting up the data and preprocessing
Starting with the transposed dataset related to the top 2000 most deviating genes calculated by MAD

```python
geneExp = "../data/E-MEXP-1304/e-mexp-1304-filtered-transposed.csv"
PyWGCNA_1304 = PyWGCNA.WGCNA(name='E-MEXP-1304-TPM5',
                                  species='arabidopsis thaliana',
                                  geneExpPath=geneExp,
                                  outputPath='/home/ribei027/wgcna/analysis_notebooks/',
                                  save=True,
                             TPMcutoff=1)
PyWGCNA_1304.geneExpr.to_df().head(5)
```

```python
PyWGCNA_1304.preprocess()
```

```python
PyWGCNA_1304.findModules()
```

## Update metadata

```python
PyWGCNA_1304.updateSampleInfo(path='../data/E-MEXP-1304/eMEXP1304SampleInfo.csv', sep=',')

# Add color to metadata

PyWGCNA_1304.setMetadataColor('Temperature', {'12': 'blue',
                                              '22': 'yellow'})

PyWGCNA_1304.setMetadataColor('Age', {'7days': 'red',
                                          '8days': 'plum',
                                          '9days': 'deeppink'})

PyWGCNA_1304.setMetadataColor('Light', {'Light': 'yellow',
                                             'Dark': 'purple'})

PyWGCNA_1304.setMetadataColor('Tissue', {'Seedling': 'green'})
```

```python
PyWGCNA_1304.analyseWGCNA()
```

```python
PyWGCNA_1304.saveWGCNA()
```

# Functional Enrichment Analysis

```python
geneList = PyWGCNA.getGeneList(dataset='athaliana_eg_gene',
                               attributes=['ensembl_gene_id',
                                           'external_gene_name',
                                           'gene_biotype'],
                               maps=['gene_id', 'gene_name', 'gene_biotype'])
```

```python
PyWGCNA_1304.updateGeneInfo(geneList)
```

```python
PyWGCNA_1304.figureType = "png"
gene_set_library = ["GO_Biological_Process_2021", "GO_Cellular_Component_2021", "GO_Molecular_Function_2021"]
PyWGCNA_1304.functional_enrichment_analysis(type="GO",
                                             moduleName="dimgrey",
                                             sets=gene_set_library,
                                             p_value=0.10,
                                             file_name="GO_Dimgrey_2021")
```

```python
PyWGCNA_1304.figureType = "png"
gene_set_library = ["KEGG_2019_Mouse"]
PyWGCNA_1304.functional_enrichment_analysis(type="KEGG",
                                             moduleName="dimgrey",
                                             sets=gene_set_library,
                                             p_value=0.05)
```

```python
PyWGCNA_1304.figureType = "pdf"
PyWGCNA_1304.functional_enrichment_analysis(type="REACTOME",
                                             moduleName="dimgrey",
                                             p_value=0.05)
```

## Module Comparison

```python
PyWGCNA_1304_TPM0 = PyWGCNA.readWGCNA("E-MEXP-1304-TPM0.p")
PyWGCNA_1304_TPM1 = PyWGCNA.readWGCNA("E-MEXP-1304-TPM1.p")
PyWGCNA_1304_TPM2 = PyWGCNA.readWGCNA("E-MEXP-1304-TPM2.p")
PyWGCNA_1304_TPM5 = PyWGCNA.readWGCNA("E-MEXP-1304-TPM5.p")
PyWGCNA_1304_TPM6 = PyWGCNA.readWGCNA("E-MEXP-1304-TPM6.p")
```

```python
comparison = PyWGCNA.compareNetworks(PyWGCNAs=[PyWGCNA_1304_TPM0,
                                                PyWGCNA_1304_TPM1,
                                                PyWGCNA_1304_TPM2,
                                                PyWGCNA_1304_TPM5,
                                                PyWGCNA_1304_TPM6])
```
